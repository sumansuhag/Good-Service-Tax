```python
import pandas as pd

# Example of train_target being a pandas Series
train_target = pd.Series([1, 2, 3, 4, 5])

# Define the MeanRegressor class
class MeanRegressor:
    def __init__(self):
        self.mean_value = None

    def fit(self, input, target):
        # Ensure target is a pandas Series or numeric array before calling .mean()
        if isinstance(target, pd.Series):
            self.mean_value = target.mean()
        else:
            raise TypeError("Target must be a pandas Series or a numeric array.")

    def predict(self, input):
        # Predict using the calculated mean value
        return [self.mean_value] * len(input)

# Instantiate and fit the model
mean_model = MeanRegressor()
mean_model.fit(train_input, train_target)

# Access the mean value
print(mean_model.mean_value)
```
